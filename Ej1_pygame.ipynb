# Move the player with the mouse button

```python
pygame.init() 

width,height = 800,400
screen = pygame.display.set_mode((width,height)) 
pygame.display.set_caption("Runner")
clock = pygame.time.Clock() 
test_font = pygame.font.Font(paths.font_pixel,50)

sky_surface = pygame.image.load(paths.img_sky).convert()
ground_surface = pygame.image.load(paths.img_ground).convert()
score_surf = test_font.render('My game',False,'black')

snail_surf = pygame.image.load(paths.img_snail1).convert_alpha()
snail_rect = snail_surf.get_rect(bottomright = (600,300))


player_surf = pygame.image.load(paths.path_Player_walk1).convert_alpha()
player_rect = player_surf.get_rect(midbottom = (80,300))

tmp = False

while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            exit() 
        if event.type == pygame.MOUSEBUTTONDOWN:
            
            if player_rect.collidepoint( pygame.mouse.get_pos() ):
                tmp = True
            
        if event.type == pygame.MOUSEBUTTONUP:
            tmp = False
            
        if event.type == pygame.MOUSEMOTION:
            
            mouse_pos = pygame.mouse.get_pos()
            
            if tmp:
                player_rect.center = mouse_pos
                
    
    screen.blit(sky_surface,(0,0))        
    screen.blit(ground_surface,(0,300))
    screen.blit(score_surf,(300,50))
    
    
    snail_rect.left -= 4
    if snail_rect.right <= 0:
        snail_rect.left = 800       
    
    screen.blit(snail_surf,snail_rect)
    
    
    #player_rect.left += 1
    screen.blit(player_surf,player_rect)
    
    
    #if player_rect.colliderect(snail_rect):
    #    print('hit')
    
    
    mouse_pos = pygame.mouse.get_pos()
    if player_rect.collidepoint( mouse_pos ):
        print('hit')
        print(pygame.mouse.get_pressed() )
    
    pygame.display.update()
    clock.tick(60) 
```

```python
pygame.init() 

width,height = 800,400
screen = pygame.display.set_mode((width,height)) 
pygame.display.set_caption("Runner")
clock = pygame.time.Clock() 
test_font = pygame.font.Font(paths.font_pixel,50)

sky_surface = pygame.image.load(paths.img_sky).convert()
ground_surface = pygame.image.load(paths.img_ground).convert()

score_surf = test_font.render('My game',False,'black')
score_rect = score_surf.get_rect(center = (400,50))

snail_surf = pygame.image.load(paths.img_snail1).convert_alpha()
snail_rect = snail_surf.get_rect(bottomright = (600,300))


player_surf = pygame.image.load(paths.path_Player_walk1).convert_alpha()
player_rect = player_surf.get_rect(midbottom = (80,300))


while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            exit() 

            
    
    screen.blit(sky_surface,(0,0))        
    screen.blit(ground_surface,(0,300))
    screen.blit(score_surf,score_rect)
    
    
    snail_rect.left -= 4
    if snail_rect.right <= 0:
        snail_rect.left = 800       
    
    screen.blit(snail_surf,snail_rect)
    
    
    player_rect.left += 1
    screen.blit(player_surf,player_rect)
    
    
    #if player_rect.colliderect(snail_rect):
    #    print('hit')
    
    
    mouse_pos = pygame.mouse.get_pos()
    if player_rect.collidepoint( mouse_pos ):
        print('hit')
        print(pygame.mouse.get_pressed() )
    
    pygame.display.update()
    clock.tick(60) 
```

```python
pygame.init() 

width,height = 800,400
screen = pygame.display.set_mode((width,height)) 
pygame.display.set_caption("Runner")
clock = pygame.time.Clock() 
test_font = pygame.font.Font(paths.font_pixel,50)
game_acive = True

sky_surface = pygame.image.load(paths.img_sky).convert()
ground_surface = pygame.image.load(paths.img_ground).convert()

score_surf = test_font.render('My game',False,'#1862f7')
score_rect = score_surf.get_rect(center = (400,50))

snail_surf = pygame.image.load(paths.img_snail1).convert_alpha()
snail_rect = snail_surf.get_rect(bottomright = (600,300))


player_surf = pygame.image.load(paths.path_Player_walk1).convert_alpha()
player_rect = player_surf.get_rect(midbottom = (80,300))
player_gravity = 0

while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            exit()
        
        if game_acive:
        
            if event.type == pygame.MOUSEBUTTONUP and player_rect.bottom >= 300:
                player_gravity = -20
            
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_SPACE and player_rect.bottom >= 300:
                    player_gravity = -20
        else:
            if event.type == pygame.KEYDOWN and event.key == pygame.K_SPACE:
                game_acive = True

    if game_acive:     
    
        screen.blit(sky_surface,(0,0))        
        screen.blit(ground_surface,(0,300))
        pygame.draw.rect(screen,'#c0e8ec',score_rect,16,7)
        screen.blit(score_surf,score_rect)
        
        snail_rect.left -= 4
        if snail_rect.right <= 0:
            snail_rect.left = 800       
        
        screen.blit(snail_surf,snail_rect)
        
        #----- Player ----- 
        player_gravity += 1
        player_rect.y += player_gravity
        if player_rect.bottom >= 300: 
            player_rect.bottom = 300
        screen.blit(player_surf,player_rect)
        
        # collision
        if snail_rect.colliderect(player_rect):
            game_acive = False  
            
    else:
        screen.fill('red')
    
    pygame.display.update()
    clock.tick(60)
```
